```python
import pandas as pd
import os 
import csv
```

```python
#Direct Method 
#school_data_load = "Resources/schools_complete.csv"
#student_data_to_load = "Resources/students_complete.csv"
#Indirecct Method
school_data_to_load = os.path.join("Resources","schools_complete.csv")
student_data_to_load = os.path.join("Resources","students_complete.csv")
```

```python
#Read the school data file and store it in a Pandas DataFrame
#Pandas uses read_csv function to read files
school_data_df = pd.read_csv(school_data_to_load)
#Can use your_data_frame.head() or .tail() to get first 5 rows or last 5 rows. Can enter specfic number want returned in parentheses
school_data_df
```

```python
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
#Counting rows in each column to make sure no null values
school_data_df.count()
```

```python
#Using isnull() function to determine if empty rows
#Null cells will say true, otherwise will be false
#Add .sum() right after to give a count of the total empty rows under each column header
student_data_df.isnull().sum()
```

```python
#Using notnull() function to determine if NOT empyt rows
#Null cells will show up as false, and ones not null will be true (opposite of isnull())
school_data_df.notnull().sum()
```

```python
school_data_df.dtypes
```

```python
prefixes_suffixes = ["Dr. ","Mr. ","Ms. ","Mrs. ","Miss "," MD"," DDS"," DVM"," PhD"]
```

```python
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
student_data_df.head(10)
```

```python
student_data_df.to_csv("student_data_cleaned_csv", index = False)
```

```python
#Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, on="school_name") #Can also do on=["school_name", "school_name"]
school_data_complete_df.head()
```

```python
#Get the total number of students
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
#Get the total number of schools
school_count = school_data_df["school_name"].count()
school_count
```

```python
#Calculate the total number of schools using the "complete" school file by only showing unique school values
school_count_2 = len(school_data_complete_df["school_name"].unique())
school_count_2
```

```python
#Calculate the total budget
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#Calculate the average reading score
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
#Calculate the average math score
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#Determine passing math and reading counts
#passing_math = school_data_complete_df["math_score"] >= 70
#passing_reading = school_data_complete_df["reading_score"] >= 70

passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
```

```python
passing_math.head()
```

```python
passing_reading
```

```python
#Calculate the number of students passing math and reading 
passing_math_count = passing_math["student_name"].count()
passing_reading_count = passing_reading["student_name"].count()
print(passing_math_count)
passing_reading_count
```

```python
#Calculate percentage that passed math and reading
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_reading_percentage = passing_reading_count / float(student_count) * 100
print(passing_math_percentage)
passing_reading_percentage
```

```python
#Calculate students who passed both math and reading 
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
passing_math_reading.head()
```

```python
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
overall_passing_math_reading_count
```

```python
overall_passing_percentage = overall_passing_math_reading_count / student_count * 100
overall_passing_percentage
```

```python
#Adding a list of values with keys to create a new DataFrame
district_summary_df = pd.DataFrame(
    [{"Total Schools": school_count, 
      "Total Students": student_count,
      "Total Budget": total_budget, 
      "Average Math Score": average_math_score,
      "Average Reading Score": average_reading_score,
      "% Passing Math": passing_math_percentage,
      "% Passing Reading": passing_reading_percentage,
      "% Overall Passing": overall_passing_percentage}])
district_summary_df
```

```python
#Format the "Total Students" to have the comma for a thousands separator
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:,.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:,.1f}".format)

district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:,.0f}%".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:,.0f}%".format)

district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:,.0f}%".format)

district_summary_df
```

```python
#Reorder the columns in the order you want them to appear
new_column_order = ["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]
#Assign a new or the same DataFrame the new column order
district_summary_df = district_summary_df[new_column_order]
district_summary_df
```
